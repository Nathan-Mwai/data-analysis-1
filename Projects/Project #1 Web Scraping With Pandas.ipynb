# 2 Read .csv from Multiple URLs with pandas

https://www.football-data.co.uk/mmz4281/2425/E0.csv

#### Link: root + season + league

```python
# link structure
"https://www.football-data.co.uk/mmz4281/" + "2425" + "/" + "E0" + ".csv"
```

```python
#Create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiple leagues

- https://www.football-data.co.uk/mmz4281/2122/E0.csv
- https://www.football-data.co.uk/mmz4281/2122/E1.csv
- https://www.football-data.co.uk/mmz4281/2122/E2.csv
- https://www.football-data.co.uk/mmz4281/2122/E3.csv
- https://www.football-data.co.uk/mmz4281/2122/EC.csv

```python
# Creating list of leagues
leagues = ['E0','E1','E2','E3','EC',]
frames = []

# Looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(f"{root}2122/{league}.csv", encoding='unicode_escape')
    frames.append(df)
# If I get Unicode Decode Error add encoding='unicode_escape' 
```

```python
# length of frames
len(frames)
```

```python
# Show 1st, 2nd and 3rd element
frames[4]
```

## 2.2 Multiple Seasons

```python
# Loop through range of seasons and insert new columns

leagues = ['E2','E3',]
frames = []

# Looping through leagues, read multiple csv and append it into a list
for league in leagues:
    for season in range(15,21):    
        df = pd.read_csv(f"{root}{season}{season+1}/{league}.csv", encoding='unicode_escape')
        df.insert(1, "season", f"20{season}")
        frames.append(df)
# If I get Unicode Decode Error add encoding='unicode_escape' 
```

```python
#Total Frames
len(frames)
```

```python
frames[11]
```

# 3. Organize All the Data in a Dictionary

```python
# Create a dictionary with original nams of the league as key
dict_countries = {
    'Spanish La Liga': 'SP1',
    'Spanish Segunda Division': 'SP2',
    'German Bundesliga':'D1',
    'English Premier League': 'E0',
    'English League 1': 'E2',
    'English League 1': 'E3',
    
}
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}



# leagues = ['E2','E3',]

# looping through element

for league in dict_countries:
    frames = []
    for season in range(15,21):    
        df = pd.read_csv(f"{root}{season}{season+1}/{dict_countries[league]}.csv", encoding='unicode_escape')
        df.insert(1, "season", f"20{season}")
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
# show dataframe inside dict_historical_data
dict_historical_data['English Premier League']
```
