```python
import pandas as pd
```

# Read a .csv from a URL With Pandas

Target website: https://www.football-data.co.uk/data.php

```python
# reading 1 csv file from a website
df_premier24 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2425/E0.csv")
```

```python
#showing the dataframe
df_premier24
#{FTHG => Final time home Goals, FTAG => FT AWAY GOALS, HTHG => Half time HG, FTR => Final time Results AKA is it home or away or draw}
```

```python
#Rename columns
df_premier24.rename(columns={
    'FTHG':"home_goals", #-> Final time home goals
    "FTAG":"away_goals" # -> final time Away goals
}, inplace=True)
```
