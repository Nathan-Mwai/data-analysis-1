```python
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# 1. Create index

```python
import numpy as np
import random
```

```python
# Create non repetitive values for the index
new_index = np.arange(0, 1000)
```

```python
#Shuffling the index
random.shuffle(new_index)
```

```python
# Create a new column with new index
df_exams['new_index'] = new_index
df_exams
```

# 2. Set index

```python
# set new_index column as index
df_exams.set_index('new_index')
```

# 3. Sort Index

```python
# Sort dataframe by new index
df_exams.sort_index()
```

```python
# Sort dataframe by new index
df_exams.sort_index(ascending=False,inplace=True)
```
