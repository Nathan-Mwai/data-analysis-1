```python
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# 1. Rename Column

```python
# rename column and overwrite values of the dataframe
df_exams = df_exams.rename(columns={
    'gender':"Gender"
})
```

```python
# rename multiple column and update the dataframe with inplace argument
df_exams.rename(columns={
    'math score':"M/S",
    'reading score':"R/S",
    "writing score":"W/S"
}, inplace=True)
```

```python
# Show dataframe
df_exams
```

# 2. Rename index 

```python
# Rename index 0, 1, 2 and updare the dataframe
df_exams.rename(index={
    0:"A",
    1:"B",
    2:"C"
}, inplace=True)
```
