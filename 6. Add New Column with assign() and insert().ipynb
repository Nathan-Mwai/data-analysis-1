# Adding a new column


## 1.1 assign()

When do I use assign?
* Adding multiple columns in a single line of code
* When I need to overwrite the values of an existing column(best practice)

It returns a new object (a copy) with all the original columns in addition to the new ones

```python
import numpy as np
```

```python
# creating random numbers for our new score columns
score1= np.random.randint(1,100,1000)
score2= np.random.randint(1,100,1000)
```

```python
# Create a series(column) using the random numbers
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
```

```python
# Using assign() to add multiple columns -> Creates a copy
df_exams.assign(score1=series1, score2=series2)
```

```python
df_exams
```

```python
#  Overwriting the copy
df_exams = df_exams.assign(score1=series1, score2=series2)
```

```python
# Shows all info in the dataframe including the update
df_exams
```

## 1.2 insert()

Insert a new column at a specific position or index Does not create a copy

```python
# using index() to add a column at a specific position
df_exams.insert(1, "test", series1)
```
